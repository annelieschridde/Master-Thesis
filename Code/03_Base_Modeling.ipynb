```python
# loading preprocessed dataset
with open("../dataframes/raw_data_Books_00_short_shuffled_lemma.pkl", "rb") as f:
    df = pickle.load(f)
```

## 1. Transformers

### 1.1 Count Vectorizer

```python
# the vectorizer object will be used to transform text to vector form
vectorizer = CountVectorizer(max_df=0.8, min_df=20, token_pattern='\w+|\$[\d\.]+|\S+')

# apply transformation
tf_fit = vectorizer.fit_transform(df['combined_clean'])
tf = tf_fit.toarray()

# tf_feature_names tells us what word each column in the matric represents
tf_feature_names = vectorizer.get_feature_names()
```

```python
tf_feature_names
```

### 1.2 TD-IDF

```python
tfidf_vec = TfidfVectorizer( 
                        max_df=0.8,
                        min_df=20,
                        token_pattern='\w+|\$[\d\.]+|\S+'
                       )
tfidf_fit = tfidf_vec.fit_transform(df['combined_clean'])
tfidf = tfidf_fit.toarray()

tfidf_feature_names = tfidf_vec.get_feature_names()
```

```python
tfidf_feature_names
```

### 1.3 Word2Vec Embeddings

```python
all_docs = []
for row in df["combined_clean"]:
    all_docs.append(row)
document = ' '.join(all_docs)
```

```python
from gensim.models import Word2Vec

sentences = nltk.sent_tokenize(document)
words =  [nltk.word_tokenize(sent) for sent in sentences]

word2vec = Word2Vec(words, min_count=2)
```

# LSA (SVD)

```python
from sklearn.decomposition import TruncatedSVD
lsa = TruncatedSVD(n_components=10, random_state=0)
```

```python
lsa_tf = lsa.fit(tf)
```

```python
def display_topics(model, feature_names, no_top_words):
    """a function which takes in our model object model, the order of the words in our matrix tf_feature_names and the number of words we would like to show. Use this function, which returns a dataframe, to show you the topics we created. Remember that each topic is a list of words/tokens and weights"""
    topic_dict = {}
    for topic_idx, topic in enumerate(model.components_):
        topic_dict["Topic %d words" % (topic_idx)]= ['{}'.format(feature_names[i])
                        for i in topic.argsort()[:-no_top_words - 1:-1]]
        topic_dict["Topic %d weights" % (topic_idx)]= ['{:.1f}'.format(topic[i])
                        for i in topic.argsort()[:-no_top_words - 1:-1]]
    return pd.DataFrame(topic_dict)
```

```python
display_topics(lsa_tf, tf_feature_names, 10)
```

```python
lsa_tfidf = lsa.fit(tfidf)
```

```python
display_topics(lsa_tfidf, tfidf_feature_names, 10)
```

# K-Means

```python
from sklearn.cluster import KMeans
k_means = KMeans(n_clusters=10, max_iter=100, random_state=0)
```

```python
def show_kmeans_results(model, nr_clusters, feature_names, nr_of_words):
    print("Top terms per cluster:")
    order_centroids = model.cluster_centers_.argsort()[:, ::-1]
    terms = feature_names
    for i in range(nr_clusters):
        print("Cluster %d:" % i),
        for ind in order_centroids[i, :nr_of_words]:
            print(' %s' % terms[ind]),
        print
```

```python
kmeans_tf = k_means.fit(tf)
labels_tf = kmeans_tf.labels_
centroids_tf = kmeans_tf.cluster_centers_
```

```python
show_kmeans_results(kmeans_tf, 10, tf_feature_names, 10)
```

```python

kmeans_tfidf = k_means.fit(tfidf)
labels_tfidf = kmeans_tfidf.labels_
centroids_tfidf = kmeans_tfidf.cluster_centers_
```

```python
show_kmeans_results(kmeans_tfidf, 10, tfidf_feature_names, 10)
```
